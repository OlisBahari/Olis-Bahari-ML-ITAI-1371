```python
# Interfacing To Library
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Separate Train and Test Data Routine
def Separate_Train_And_Test_Data(X, y):

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.3,
        random_state=42
    )

    return X_train, X_test, y_train, y_test
```

```python

# Load Iris Dataset Routine
def Load_Iris_Dataset():

    # Load the Iris dataset
    iris = load_iris()

    # Separate features and target
    X = iris.data
    y = iris.target

    return iris, X, y

    
```

```python
# Train Single Decision Tree Routine
def Train_Single_Decision_Tree(X_train, X_test, y_train, y_test):

    # Create the Decision Tree model
    decision_tree = DecisionTreeClassifier(random_state=42)

    # Train the model
    decision_tree.fit(X_train, y_train)

    # Make predictions
    y_pred_tree = decision_tree.predict(X_test)

    # Evaluate accuracy
    accuracy_tree = accuracy_score(y_test, y_pred_tree)

    return decision_tree, accuracy_tree
```

```python
# Train Random Forest Routine
def Train_Random_Forest(X_train, X_test, y_train, y_test):

    # Create the Random Forest model
    random_forest = RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )

    # Train the model
    random_forest.fit(X_train, y_train)

    # Make predictions
    y_pred_rf = random_forest.predict(X_test)

    # Evaluate accuracy
    accuracy_rf = accuracy_score(y_test, y_pred_rf)

    return random_forest, accuracy_rf
```

```python
# Plot Random Forest Feature Importance Routine
def Plot_Random_Forest_Feature_Importance(random_forest, feature_names):

    # Get feature importance values
    importances = random_forest.feature_importances_

    # Create the feature importance graph
    plt.figure(figsize=(8, 5))
    plt.bar(feature_names, importances)

    plt.title("Feature Importance - Random Forest")
    plt.xlabel("Features")
    plt.ylabel("Importance")
    plt.xticks(rotation=20)
    plt.tight_layout()

    plt.show()
```

```python
# <<< Main Routine >>>
def Main():

    # Load the dataset
    iris, X, y = Load_Iris_Dataset()

    # Separate training and testing data
    X_train, X_test, y_train, y_test = (Separate_Train_And_Test_Data(X, y))

    # Train and evaluate the Decision Tree
    _, accuracy_tree = Train_Single_Decision_Tree(X_train,X_test,y_train,y_test)

    # Train and evaluate the Random Forest
    random_forest, accuracy_rf = Train_Random_Forest(X_train,X_test,y_train,y_test)

    # Plot Random Forest feature importances
    Plot_Random_Forest_Feature_Importance(random_forest,iris.feature_names)
 
    print("\nModel Comparison")
    print(f"Decision Tree Accuracy: {accuracy_tree:.2%}")
    print(f"Random Forest Accuracy: {accuracy_rf:.2%}")
    
```

```python
# Run the program
if __name__ == "__main__":
    Main()
```
