```python
# Interfacing To Library
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Seprate Train And Test Data Routine
def Seprate_Train_And_Test_Data():

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    return X_train, X_test, y_train, y_test
```

```python
# Load Iris Dataset Routine
def Load_Iris_Dataset():
    
    # Load and prepare data
    iris = load_iris()
    X = iris.data
    y = iris.target

    X_train, X_test, y_train, y_test = Seprate_Train_And_Test_Data()

    return X, y

    
```

```python
# Train Single Decision Tree Routine
def Train_Single_Decision_Tree():
    # --- Part 3: Single Decision Tree ---

    # Create and train a Decision Tree
    decision_tree = DecisionTreeClassifier(random_state=42)
    decision_tree.fit(X_train, y_train)

    # Make predictions
    y_pred_tree = decision_tree.predict(X_test)

    # Evaluate accuracy
    accuracy_tree = accuracy_score(y_test, y_pred_tree)

    print(f"Accuracy of the Decision Tree: {accuracy_tree:.2%}")
```

```python
# Traiin Random Forest Routine
def Train_Random_Forest():
    # --- Part 4: Random Forest ---

    # Create and train a Random Forest
    random_forest = RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )

    random_forest.fit(X_train, y_train)

    # Make predictions
    y_pred_rf = random_forest.predict(X_test)

    # Evaluate accuracy
    accuracy_rf = accuracy_score(y_test, y_pred_rf)

    print(f"Accuracy of the Random Forest: {accuracy_rf:.2%}")
```

```python
# Plot Random Forest Feature Importance Routine
def Plot_Random_Forest_Feature_Importance():

    # Get feature importances
    importances = random_forest.feature_importances_

    # Feature names
    feature_names = iris.feature_names

    # Plot feature importance
    plt.figure(figsize=(8,5))
    plt.bar(feature_names, importances)
    plt.title("Feature Importance - Random Forest")
    plt.xlabel("Features")
    plt.ylabel("Importance")
    plt.xticks(rotation=20)

    plt.show()
```

```python
# <<< Main Routine >>>

Load_Iris_Dataset()

Train_Single_Decision_Tree()

Train_Random_Forest()

Plot_Random_Forest_Feature_Importance()
```
