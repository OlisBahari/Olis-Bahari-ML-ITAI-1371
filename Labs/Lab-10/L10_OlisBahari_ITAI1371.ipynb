```python
import warnings
warnings.filterwarnings(
    "ignore",
    message="KMeans is known to have a memory leak on Windows with MKL.*"
)

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.datasets import make_blobs, load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
def Generate_Synthetic_Data():
    # Generate synthetic data
    X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=42)
    
    # Visualize the data
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], s=50)
    plt.title("Synthetic Customer Data")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
def K_Means_Cluster_Elbow_Method_Model():

    # Create an empty list to store inertia values
    inertia = []

    # Test values of k from 1 through 10
    for k in range(1, 11):

        # Create the K-Means model
        kmeans = KMeans(
            n_clusters=k,
            random_state=42,
            n_init=10
        )

        # Train the model
        kmeans.fit(X)

        # Store the model's inertia
        inertia.append(kmeans.inertia_)

    # Plot the elbow curve
    plt.figure(figsize=(10, 6))

    plt.plot(
        range(1, 11),
        inertia,
        marker="o"
    )

    plt.title("Elbow Method for Selecting k")
    plt.xlabel("Number of Clusters (k)")
    plt.ylabel("Inertia")
    plt.xticks(range(1, 11))
    plt.grid(True)
    plt.show()
```

```python
def K_Means_Clustering_Model():

    # Create and train the K-Means model
    kmeans = KMeans(
        n_clusters=4,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X)

    # Get the cluster assigned to each point
    labels = kmeans.labels_

    # Get the coordinates of the cluster centers
    centers = kmeans.cluster_centers_


    # Plot the clusters
    plt.figure(figsize=(10, 6))

    plt.scatter(
        X[:, 0],
        X[:, 1],
        c=labels,
        s=50,
        cmap="viridis"
    )

    # Plot the cluster centers
    plt.scatter(
        centers[:, 0],
        centers[:, 1],
        c="red",
        s=200,
        alpha=0.75,
        marker="X",
        label="Cluster Centers"
    )

    plt.title("K-Means Clustering Results (k = 4)")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```

```python
def PCA_Model():

    # Load the Iris dataset
    iris = load_iris()

    X_iris = iris.data
    y_iris = iris.target


    # Create a PCA model that keeps two principal components
    pca = PCA(n_components=2)

    # Fit PCA and transform the Iris data
    X_pca = pca.fit_transform(X_iris)


    # Display the amount of variance explained
    print(
        "Explained variance by component:",
        pca.explained_variance_ratio_
    )

    print(
        f"Total variance explained by 2 components: "
        f"{pca.explained_variance_ratio_.sum():.2%}"
    )


    # Create a DataFrame for visualization
    pca_df = pd.DataFrame(
        data=X_pca,
        columns=[
            "Principal Component 1",
            "Principal Component 2"
        ]
    )

    # Use the species names instead of the numbers 0, 1, and 2
    pca_df["Species"] = [
        iris.target_names[target]
        for target in y_iris
    ]


    # Plot the PCA results
    plt.figure(figsize=(10, 7))

    sns.scatterplot(
        x="Principal Component 1",
        y="Principal Component 2",
        hue="Species",
        data=pca_df,
        palette="viridis",
        s=100
    )

    plt.title("PCA of Iris Dataset — 2 Components")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.legend(title="Species")
    plt.show()
```

```python
# <<<< Main >>>

Generate_Synthetic_Data()

K_Means_Cluster_Elbow_Method_Model()

K_Means_Clustering_Model()

PCA_Model()
```

[Q1]

Supervised learning uses labeled data, where the correct output, or target, is already known. Its main point is to learn the relationship between the input features and the target so it can make predictions on new data.

Unsupervised learning uses unlabeled data, where no target values are provided. Its main point is to find patterns without target labels. 

Unsupervised learning identifies hidden patterns, relationships, or groups in unlabeled data, such as clusters produced by the K-Means algorithm. Its main point is to uncover structure in the data.

[Q2]

Increasing k always reduces inertia, but choosing the largest possible k is not useful because the model begins to overfit the data. When k equals the number of data points, each point becomes its own cluster, resulting in nearly zero inertia but providing no meaningful grouping. The elbow method helps identify the point at which adding more clusters yields only small improvements, so the main point is to balance simplicity and accuracy.

[Q3]

If the first two principal components explain more than 95% of the variance, most of the important information in the original 4-dimensional dataset can be captured in only two dimensions. This indicates that the four original features are highly correlated, allowing PCA to reduce the dataset's dimensionality with minimal information loss. As a result, the Iris dataset can be visualized effectively in two dimensions while preserving most of its underlying structure. The main point is that PCA can simplify the dataset without losing much information.
